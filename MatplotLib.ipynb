## Import Libraries

```python
"""
Matplotlib official website
https://matplotlib.org/
"""
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Basics

```python
x= np.arange(0,5,0.001) # The more values the more detailed curve
y = np.sin(x)
z = np.cos(x)
# plt.plot(x,y,label='2x',color='red',linewidth=2, marker='*',markersize=10,markeredgecolor='#125794')

plt.plot(x,y,'b',label='sinx',linestyle='dashdot')
plt.plot(x,np.cos(x),'r',label='cosx',linestyle='dashdot')
# plt.plot(x,np.tan(x),'y',label='tanx')
plt.title("Basic Graph", fontdict={'fontsize':20})

plt.xlabel("X",fontdict={'fontsize':20})
plt.ylabel("Y",fontdict={'fontsize':20})

plt.legend()

plt.show()
```

## Pie Charts

```python
import random

def different_color_generator(n:int) -> list:
    colors = [] # list to store the random rgba colors
    for i in range(n):
        # RGBA values will be between 0 and 1
        # append it to the list colors
        colors.append((random.uniform(0,random.uniform(0,1)), 
                       random.uniform(0,random.uniform(0,1)), 
                       random.uniform(0,random.uniform(0,1))
                      ))
    return colors # Return colors list

pie = plt.figure(figsize=(6,4) , dpi=100)
pie_axes = pie.add_axes([0.1,.1,.9,.9])

# Random words
sample = """marvelous
rotten
airport
strengthen
rush
stroke
delightful
record
slippery
vulgar
needless
sticks
lighten
feigned
obey
""".split("\n")
# Generating random values for the words
vals = [random.randrange(100,500) for _ in range(len(sample))]
# Generate the color for the samples length
colors = different_color_generator(len(sample)) 
# Explode for the length of the samples
explode = [0] * len(sample)
# Change the values so that it can explode outside the pie chart
explode[random.randrange(0,len(sample))] = 0.2
# ploting the pie chart
wedges,txt,autotxt = plt.pie(
    vals, # The values of random generated number
    explode=explode, # exploding list
    labels=sample, # Labels as samples (random words)
    colors=colors, # Colors list
    autopct="%1.0f%%", # The floating point 
    shadow=True,# Shadow of the pie chart
    startangle=160, # Starting angle of the pie chart
    textprops=dict(color='w') # text props of type dictionary 
)
plt.legend(
    wedges, # Wedges 
    sample, # The sample 
    bbox_to_anchor=(1,0,0.5,1) # Where should be the box should be placed
)
```

## Heat maps

```python
def format_for_heat_map(arr, n) :
    if arr.size%n != 0: # If the lenght of the file is not divisible by n
        raise ValueError(
        "We cannnot split the array equaly"
        )
    formated_arr = [] # Formated Array where the splited array will be stored
    for i in range(arr.size//n): # Iterate over the array size by n
        formated_arr.append(arr[i:i+n]) # Append the splited array to the formatted _arr
 
    return np.array(formated_arr) # Convert the result to numpy array then return it
countries = countries[0:5] # Get first 5 values of the countries

deaths_last_24 = corona_data_arr[0:,-1].astype(int) / 100 # Get the last col of the array then convert to int
deaths_last_24 = format_for_heat_map(
    deaths_last_24,
    countries.size
) # Format the 1d array to 2d array
FORMAT = "%b %d %Y" # Format of the datetime
dates = np.array(
    [ datetime.strptime(f"Jun {str(i)} 2021",FORMAT) for i in range(25,30) ]
) # Generate the dates 
dates = np.array(
    [datetime.strftime(i,FORMAT[0:5]) for i in dates]
) # Convert the datetime to string 

heat_map_fig, heat_map_axes = plt.subplots() # Create a subplots
im = heat_map_axes.imshow(
    deaths_last_24, # To plot the past deaths in 24 hrs
    cmap=cmaps_arr[np.random.randint(0, len(cmaps_arr) - 1)] # Color maps
) # Show the image 
heat_map_axes.set_xticks(np.arange(len(dates))) # Set the x ticks  
heat_map_axes.set_yticks(np.arange(len(deaths_last_24))) # Set the y ticks
heat_map_axes.set_xticklabels(dates) # Set the x tick label with dates 
heat_map_axes.set_yticklabels(countries) # Set the y tick label with countries

plt.setp( # Setting position of the x tick labels
    heat_map_axes.get_xticklabels() , # Get the x tick labels
    rotation=45, # Rotation degrees
    ha="right", # Horizontal alignment
    rotation_mode="anchor", # Rotation mode
    va="top"  # Vertical alignment
)
plt.setp( # Setting position for the y tick labels
    heat_map_axes.get_yticklabels(), # Get the y tick labels
    rotation=45, # Rotation degrees
    ha="right", # Horizontal alignment
    rotation_mode="anchor", # Rotation mode
    va="bottom" # Vertical alignment
)
for i in range(len(countries)): 
    for j in range(len(dates)):
        text = heat_map_axes.text(
            j, # X axis
            i, # Y axis
            deaths_last_24[i,j], # The data
            ha="center", # Horizontal alignment
            va="center", # Vertical alignment
            color="k", # Black color
            fontweight="bold" # Font weight of the text
        )

plt.show() # Show the figures
```
