## Import Libraries

```python
"""
Matplotlib official website
https://matplotlib.org/
"""
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Basics

```python
x= np.arange(0,5,0.001) # The more values the more detailed curve
y = np.sin(x) # convert the array x to sin(x)
z = np.cos(x) # Convert the array x to cos(x)
# plt.plot(x,y,label='2x',color='red',linewidth=2, marker='*',markersize=10,markeredgecolor='#125794')

plt.plot(x,y,'b',label='sinx',linestyle='dashdot') # First plot x and y axis
plt.plot(x,z,'r',label='cosx',linestyle='dashdot') # Second plot x and z axis
plt.title("Basic Graph", fontdict={'fontsize':20}) # Title of the graph

plt.xlabel("X",fontdict={'fontsize':20}) # X axis label with custom font size
plt.ylabel("Y",fontdict={'fontsize':20}) # Y axis label with custom font size

plt.legend() # Plot the legend

plt.savefig("a.jpeg",dpi=100)
plt.show() # Plot the graph
```

## Predicting Other Values

```python
plt.plot(
    x[:y.size//2],
    y[:y.size // 2],
    label='sinx',
    c='r'
) # Plot the first half values with normal linestyle
plt.plot(
    x[y.size//2:],
    y[y.size//2:],
    linestyle='dotted',
    c='b',
    linewidth=3
) # Plot the second half values with dotted
plt.show() # Show the graph
```

## Multiple Plots


```python
plt.subplot(1,2,1) # Create a sub plots with 1 row , 2 cols , choosing the current col as 1
plt.plot(x,y,'r') # Plot x and y with color red 
plt.subplot(1,2,2) # Create a sub plot with 1 row , 2 cols choosing the current col as 2
plt.plot(x,z,'b') # Plot x and z with color blue
plt.show() # Plot the graph
```

## Figure

```python
figure_one = plt.figure( # Create a figure object
    figsize=(5,4),  # Figure size 
    dpi=100 # Dots per inch
)
axes_one = figure_one.add_axes( # Add axes 
    [0.1,0.1,0.9,0.9] # [left, top, right, bottom]
)
axes_one.set_xlabel("x") # Set x label
axes_one.set_ylabel("Sin(x)") # Set y axis label
axes_one.set_title("Axis 1") # Set title of the axis 1
axes_one.plot(
    x, # X data
    y, # Y data
    label="sin(x)" # Label 
) 
axes_one.plot(
    y, # X data
    x, # Y data
    label="x" # Label
)
axes_one.legend() # Create and show the legend
axes_two = figure_one.add_axes( # Add axes 
    [0.5,0.5,0.4,0.3]  # [left, top, right, bottom]
)

axes_two.set_xlabel("x")  # Set x label
axes_two.set_ylabel("cos(x)") # Set y axis label
axes_two.set_title("Axis 2") # Set title of the axis 2
axes_two.plot( # Plot for the inner axis
    x, # X data
    z, # Y data
    label="cos(x)" # Label
)
axes_two.plot( # Plot for the inner axis
    z, # X data
    x, # Y data
    label="x" # Label
)
axes_two.text(
    2, # x axis
    2, # y axis
    "Hello" # Text to be printed at (x,y)
)
plt.show() # Show the graph
 
```

## Subplots

```python
figure,axes = plt.subplots(
    figsize = (8,4), # Figure size 
    nrows = 1, # Number of rows
    ncols = 3  # Number of cols
)
plt.tight_layout() # Set the layout to tight layout
length, k = x.size//3, 0 
for i in range(3): # Iterate over 3 plots (ncols)
    axes[i].set_title(f"Plot {i+1}") # Set the title of the ith plot  
    axes[i].set_xlabel("X") # Set the x axis label of the ith plot
    axes[i].set_ylabel("sin(X)") # Set the y axis label of the ith plot
    axes[i].plot( # plot the graph of the ith plot 
        x[k:k+length], # Plot the x data
        y[k:k+length] # plot the y data
    ) 
    k+=length
plt.show() # Plot the graphs
```

## Pie Charts

```python
import random

def different_color_generator(n:int) -> list:
    colors = [] # list to store the random rgba colors
    for i in range(n):
        # RGBA values will be between 0 and 1
        # append it to the list colors
        colors.append((random.uniform(0,random.uniform(0,1)), 
                       random.uniform(0,random.uniform(0,1)), 
                       random.uniform(0,random.uniform(0,1))
                      ))
    return colors # Return colors list

pie = plt.figure(figsize=(6,4) , dpi=100)
pie_axes = pie.add_axes([0.1,.1,.9,.9])

# Random words
sample = """marvelous
rotten
airport
strengthen
rush
stroke
delightful
record
slippery
vulgar
needless
sticks
lighten
feigned
obey
""".split("\n")
# Generating random values for the words
vals = [random.randrange(100,500) for _ in range(len(sample))]
# Generate the color for the samples length
colors = different_color_generator(len(sample)) 
# Explode for the length of the samples
explode = [0] * len(sample)
# Change the values so that it can explode outside the pie chart
explode[random.randrange(0,len(sample))] = 0.2
# ploting the pie chart
wedges,txt,autotxt = plt.pie(
    vals, # The values of random generated number
    explode=explode, # exploding list
    labels=sample, # Labels as samples (random words)
    colors=colors, # Colors list
    autopct="%1.0f%%", # The floating point 
    shadow=True,# Shadow of the pie chart
    startangle=160, # Starting angle of the pie chart
    textprops=dict(color='w') # text props of type dictionary 
)
plt.legend(
    wedges, # Wedges 
    sample, # The sample 
    bbox_to_anchor=(1,0,0.5,1) # Where should be the box should be placed
)
```
